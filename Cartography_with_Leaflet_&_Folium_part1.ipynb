# Cartography with Leaflet & Folium - Part 1
**Cartography** is the study and practice of making **maps**. Combining **science**, **aesthetics**, and technique, cartography builds on the premise that reality (or an imagined reality) can be modeled in ways that communicate spatial information effectively.  

The fundamental uses of traditional cartography are to:  
- Set the map's agenda and select traits of the object to be mapped. This is the concern of map editing. Traits may be physical, such as roads or land masses, or may be abstract, such as toponyms or political boundaries.
- Represent the terrain of the mapped object on flat media. This is the concern of map projections.
- Eliminate characteristics of the mapped object that are not relevant to the map's purpose. This is the concern of generalization.
- Reduce the complexity of the characteristics that will be mapped. This is also the concern of generalization.
- Orchestrate the elements of the map to best convey its message to its audience. This is the concern of map design.  
Source : [Wikipedia](https://en.wikipedia.org/wiki/Cartography)

## Folium library
The JavaScipt library **Leaflet** is available in the **folium** library for the Python environment.  
`folium` is an-easy-to-use library, instructions are clear and concise.  Maps are created by adding multiple layers of information corresponding to your needs.  
The official documentation is available here : [folium](https://python-visualization.github.io/folium/modules.html).

```
import folium
```

## Create a first map
In this first example, we declare the following instructions :
- Definition of geographical positions, **latitude** (`lat`) & **longitude** (`lon`),
- Creation of a map nammed `map` centered on the previous positions with a zoom level of 12 (index of zoom level),
- Calling the map for rendering on the screen. 

```
lat = 47.3931281
lon = 0.6687137

map = folium.Map(location = [lat, lon], zoom_start = 14)
map
```

## Customize the map
**Leaflet** is an extremely rich library that offers many possibilities for making a beautiful map, such as :
- Change the size of the map
- Add markers
- Add circles
- Add popups
- Add tooltips
- Change the tiles
- and so more other features described in the [Leaflet' official documentation](https://leafletjs.com/).

##Change the size
To make easier the readability of the maps, we decrease a bit the default size of the figures by adjusting both `width` and `height` of the `.Figure()` method.  
Then the parameters is applied to the map with the `add_to()` method.

```
fig = folium.Figure(width=800, height=500)
map = folium.Map(location = [lat, lon], zoom_start = 14).add_to(fig)
map
```

##Add markers
### A simple example
Marker is often the first tool to customize.  
Here we add a simple marker located at the previous positions with the method `.Marker()` and the popup message is set to "MAME".  
Marker is simply added to the previous map with the method `.add_to()`.


```
folium.Marker(location = [lat, lon],  popup = "MAME").add_to(map)
map
```

### A more complexe example
Here we customise the messages shown when cliking and hovering the marker.  
An icon is added to the marker as well.

```
fig = folium.Figure(width=800, height=500)
map = folium.Map(location = [lat, lon], zoom_start = 14).add_to(fig)
folium.Marker(location = [lat, lon], 
              popup = "A message here when <font color=blue><strong>clicking</strong></font> on the marker",
              tooltip="A message here when <font color=orange><strong>hovering</strong></font> the marker",
              icon=folium.Icon(color='green', icon='home', prefix='fa'),
              ).add_to(map)
map
```

##Add circles
### A simple example
Circle is often use for highlighting area of interest.  
Here we add a simple circle located at the previous positions with the method `.Circle()` and the popup message is set to "MAME".  
Circle is simply added to the previous map with the method `.add_to()`.

```
fig = folium.Figure(width=800, height=500)
map = folium.Map(location = [lat, lon], zoom_start = 14).add_to(fig)
folium.Circle(location = [lat, lon],  popup = "MAME", radius=100).add_to(map)
map
```

### A more complexe example
Here we customise the messages shown when cliking and hovering the circle and colors for the circle as well.

```
fig = folium.Figure(width=800, height=500)
map = folium.Map(location = [lat, lon], zoom_start = 14).add_to(fig)
folium.Circle(location = [lat, lon], 
              popup = "A message here when <font color=blue><strong>clicking</strong></font> on the circle",
              tooltip="A message here when <font color=orange><strong>hovering</strong></font> the circle",
              radius=250,
              color='crimson',
              fill=True,
              fill_color='crimson'
              ).add_to(map)
map
```

##Add popups
We have seen the popup features added with markers and circles.  
Popups can be customized with the `.Popup()` method.

```
fig = folium.Figure(width=800, height=500)

popup = folium.Popup("Popup : A message here when <font color=blue><strong>clicking</strong></font> on the marker.<br>MAME area, here I go for learning data !<br>And I love it :)",max_width=500)

map = folium.Map(location = [lat, lon], zoom_start = 14).add_to(fig)
folium.Marker(location = [lat, lon], 
              popup = popup,
              tooltip="A message here when <font color=orange><strong>hovering</strong></font> the marker",
              icon=folium.Icon(color='green', icon='home', prefix='fa'),
              ).add_to(map)
map
```

##Add tooltips
We have seen the tooltip features added with markers and circles.  
Tooltips can be customized with the `.Tooltip()` method.

```
fig = folium.Figure(width=800, height=500)

tooltip = folium.Tooltip(text="Tooltip : A message here when <font color=orange><strong>hovering</strong></font> on the marker.<br>MAME area, here I go for learning data !<br>And I love it :)")

map = folium.Map(location = [lat, lon], zoom_start = 14).add_to(fig)
folium.Marker(location = [lat, lon], 
              popup = popup,
              tooltip=tooltip,
              icon=folium.Icon(color='green', icon='home', prefix='fa')
              ).add_to(map)
map
```

## Change the tiles
The tiles are the basically the background of the map.  
Sereval providers are available offering different kind of aesthetics, complementary information, etc.  
- Open Stree tMap is the default tiles
- CartoDB Positron
- Stamen Terrain
- etc.


### A new tiles
When calling a map, the `tiles` parameter can be set at this time.  
The example below creates a map with the CartoDB Positron tiles.


```
fig = folium.Figure(width=800, height=500)
map = folium.Map(location = [lat, lon], tiles='cartodbpositron', zoom_start = 14).add_to(fig)
map
```

###The `.TileLayer()` feature
The `.TileLayer()` feature offers the ability to store several tiles for the map and let the possibility to user of switching to different tiles.

```
fig = folium.Figure(width=800, height=500)
map = folium.Map(location = [lat, lon], tiles='OpenStreetMap', zoom_start = 14).add_to(fig)

folium.TileLayer('cartodbpositron', name="CartoDB Positron").add_to(map)
folium.TileLayer('cartodbdark_matter', name="CartoDB Dark Matter").add_to(map)
folium.TileLayer('stamenterrain', name="Stamen Terrain").add_to(map)
folium.LayerControl().add_to(map)
map
```

## Custom map with your own data


### Create a dataframe

```
import folium
import pandas as pd
 
# A Pandas dataframe with your own data
data = pd.DataFrame({
   'lat':[47.394622, 47.394424, 47.39443, 47.39336],
   'lon':[0.681866, 0.681486, 0.682542, 0.681265],
   'name':['Au Temps des Rois', 'Le Vieux Murier', 'Les Trois Orfevres', 'Le Strapontin'],
   'rate':[4.9, 3.9, 4.3, 5.0]
})

data
```

###Adapt data for the map
Here we convert quantitative data `rate` into qualitative data `color` for displaying icons on the map based on the rate data.


```
data['color'] = pd.cut(data['rate'], bins=[0, 3.5, 4.0, 4.5, 5], labels=["brown", "red", "orange", "green"])
data
```

### Make a map 

```
fig = folium.Figure(width=800, height=500)

map = folium.Map(location = [lat, lon], zoom_start = 14).add_to(fig)
map
```

### Fill the map with the data

```
# Fill the map with your own dataframe
for lat, lon, name, rate, color in zip(data['lat'], data['lon'],  data['name'], data['rate'], data['color']):
  folium.Marker(location=[lat, lon],
                popup=name,
                tooltip=str("Name : " + str(name)+ "<br>Rate : "+ str(rate)),
                icon=folium.Icon(color=color, icon='home', prefix='fa')
                ).add_to(map)

map
```

## Wrap up
In the following example we wrap all previous features up and make an interactive Leaflet' map.  

```
#Step 1 : definition of map and tiles
fig = folium.Figure(width=800, height=500)
map = folium.Map(location = [lat, lon], zoom_start = 14).add_to(fig)
folium.TileLayer('cartodbpositron', name="CartoDB Positron").add_to(map)
folium.TileLayer('cartodbdark_matter', name="CartoDB Dark Matter").add_to(map)
folium.TileLayer('stamenterrain', name="Stamen Terrain").add_to(map)
folium.LayerControl().add_to(map)

# Step 2 : add marker where we love to learn data
popup_marker = folium.Popup("MAME area, here I go for <strong>learning data</strong> !<br>And I love it :)", max_width=500)
tooltip_marker = folium.Tooltip(text="MAME area, lot of fun here!")
lat, lon = 47.3931281, 0.668713
folium.Marker(location = [lat, lon], 
              popup=popup_marker,
              tooltip=tooltip_marker,
              icon=folium.Icon(color='blue', icon='home', prefix='fa'),
              ).add_to(map)

# Step 3 : add circle for the area I love to go
popup_circle = folium.Popup("Tours Center area, here I go for drinking a beer<br>when data is aweful with me :( !", max_width=500)
tooltip_circle = folium.Tooltip(text="Tours Center area, lot of relaxing places here!")
folium.Circle(location = [47.394421, 0.681727], 
              popup=popup_circle,
              tooltip=tooltip_circle,
              radius=500,
              color='crimson',
              fill=True,
              fill_color='crimson'
              ).add_to(map)

# Step 4 : fill the map with your own dataframe
data = pd.DataFrame({
   'lat':[47.394622, 47.394424, 47.39443, 47.39336],
   'lon':[0.681866, 0.681486, 0.682542, 0.681265],
   'name':['Au Temps des Rois', 'Le Vieux Murier', 'Les Trois Orfevres', 'Le Strapontin'],
   'rate':[4.9, 3.9, 4.3, 5.0]
})
data['color'] = pd.cut(data['rate'], bins=[0, 3.5, 4.0, 4.5, 5], labels=["brown", "red", "orange", "green"])

for lat, lon, name, rate, color in zip(data['lat'], data['lon'],  data['name'], data['rate'], data['color']):
  folium.Marker(location=[lat, lon],
                popup=name,
                tooltip=str("Name : " + str(name)+ "<br>Rate : "+ str(rate)),
                icon=folium.Icon(color=color, icon='home', prefix='fa')
                ).add_to(map)

# Step 5 : render the final map !
map
```
