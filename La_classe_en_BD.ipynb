>Les classes sont un moyen de réunir des données et des fonctionnalités.  
Créer une nouvelle classe crée un nouveau type d'objet et ainsi de nouvelles instances de ce type peuvent être construites.  
Chaque instance peut avoir ses propres attributs, ce qui définit son état.  
Une instance peut aussi avoir des méthodes (définies par la classe de l'instance) pour modifier son état.
[python.org](https://docs.python.org/fr/3/tutorial/classes.html)

### Introduction
Imaginons que nous voulions réaliser la classification  des animaux sous forme d'un [arbre phylogénétique](https://fr.wikipedia.org/wiki/Classification_phylog%C3%A9n%C3%A9tique).  
Cela pourrait ressembler à ça :
Classification des animaux

Nous pourrions alors adopter le système de `class` de Python, et cela pourrait ressembler à ça ...

```
class Animal: # Définition de la classe Animal
    """Un 'docstring' ici donnant les explications de notre classe : 
    classe Animal caractérisée par la présence
    - d'un nom
    - d'une tête
    - et/ou d'une bouche
    - et/ou des yeux"""
    
    
    def __init__(self, nom): # Le constructeur de notre classe
        """Les attributs de l'animal"""
        self.nom = nom
        self.tete = True
        self.bouche = True
        self.oeil = True

    def __str__(self):
        """Méthode appelée lors d'un appel de type print (conversion de l'objet en chaîne de caracteres)"""
        return "{0} est mon animal de compagnie. Grâce à lui j'ai le droit de sortir de chez moi en ce moment !".format(self.nom)
```

### La classe vertébré est rattachée à la classe animal

Reprenons notre classification animale pour créer les classes appartenant à la classe `Animal()` : la classe `Vertebre()` est directement rattachée à la classe `Animal()`.  
Il s'agit de créer des classes **filles** appartenant à la classe **mère** et de transmettre un **héritage** d'**attibuts**, de **méthodes**, etc.  
La syntaxe est relativement simple et intuitive : `class ClasseFille(ClasseMere):`

```
class Vertebre(Animal):
    """Classe Vertebre, qui hérite de Animal.
    Elle reprend les mêmes méthodes et attributs de la classe mère
    Ici aucune méthode ni attribupour commencer ... """
```

Nous créons ici un nouvel animal de la classe `Vertebre()`...

```
mickey = Vertebre("MickeyMouse")
```

... qui hérite des attributs et méthodes de la classe mère `Animal()` :

```
mickey.bouche
```

```
print(mickey)
```

Reprenons la classe `Vertebre()` pour la compléter et l'améliorer ...

```
class Vertebre(Animal):
    """Classe Vertebre, qui hérite de Animal.
    Elle reprend les mêmes méthodes et attributs de la classe mère"""


    def __init__(self, nom):
        """Selon la classification des animaux, un vertébér a aussi un squellete et un crane"""
        Animal.__init__(self, nom)
        self.squellette = True
        self.crane = True
```

```
mickey = Vertebre("MickeyMouse")
mickey.squellette
```

```
mickey.crane
```

Vérifions également que la classe `Vertebre()` a bien hérité des méthodes et attributs de sa classe mère `Animal()` :

```
mickey.tete
```

```
print(mickey)
```

### La classe mamifère est rattachée à la classe vertébré
Rantanplan est bien un animal, mais il serait plus juste de le classer parmis les mamifères. Créons cette nouvelle classe qui hérite de sa classe mère `Vertebre()` :

```
class Mamifere(Vertebre):
    """Classe Mamifere, qui hérite de Vertebre.
    Elle reprend les mêmes méthodes et attributs de sa classe mère"""


    def __init__(self, nom):
        """Selon la classification des animaux, un mamifère a aussi 4 membres, des poils et des mamelles"""
        Vertebre.__init__(self, nom)
        self.membres = 4
        self.poils = True
        self.mamelles = True
```

```
# Rantanplan est maintenant un mamifère, il n'est plus dans la classe animal
rantanplan = Mamifere("Rantanplan")
```

```
rantanplan
```

### Pour conclure
* On définit une classe en suivant la syntaxe `class NomClasse:`.
*   Les méthodes se définissent comme des fonctions, sauf qu'elles se trouvent dans le corps de la classe.
* Les méthodes d'instance prennent en premier paramètre `self`, l'instance de l'objet manipulé.
* On construit une instance de classe en appelant son constructeur, une méthode d'instance appelée `__init__`.
* On définit les attributs d'une instance dans le constructeur de sa classe, en suivant cette syntaxe : `self.nom_attribut = valeur`.
